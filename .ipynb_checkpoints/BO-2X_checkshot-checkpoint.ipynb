## Smoothing and de-spiking of logs

```python
def rolling_window(a, window):
        shape = a.shape[:-1] + (a.shape[-1] - window + 1, window)
        strides = a.strides + (a.strides[-1],)
        rolled = np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)
        return rolled
    
window = 13 # the length of filter is 13 samples or ~ 2 metres
rho_sm = np.median(rolling_window(bo2.rho,window), -1)
rho_sm = np.pad(rho_sm, np.int(window/2), mode='edge')

plt.figure(figsize=(10,4))
plt.plot(bo2.depth,rho_sm,'b', bo2.depth, bo2.rho, 'k', alpha=0.5)
plt.plot(bo2.depth, rho_sm)
plt.show()

def despike(curve, curve_sm, max_clip): 
    spikes = np.where(curve - curve_sm > max_clip)[0]
    spukes = np.where(curve_sm - curve > max_clip)[0]
    out = np.copy(curve)
    out[spikes] = curve_sm[spikes] + max_clip  # Clip at the max allowed diff
    out[spukes] = curve_sm[spukes] - max_clip  # Clip at the min allowed diff
    return out

rho = despike(bo2.rho, rho_sm, max_clip = 100)

plt.figure(figsize=(10,4))
plt.plot(bo2.depth, bo2.rho,'k')
plt.plot(bo2.depth, rho_sm,'b')
plt.plot(bo2.depth, rho,'r')
plt.title('de-spiked density')

dt_sm = np.median(rolling_window(bo2.vp,window), -1)
dt_sm = np.pad(dt_sm, np.int(window/2), mode='edge')
dt = despike(bo2.vp, dt_sm, max_clip = 10)

ind_ = np.nonzero(rho) #locate all non-zero indices in rho-array

plt.figure(figsize=(10,4))
plt.plot(bo2.depth, bo2.vp,'k')
plt.plot(bo2.depth, dt_sm,'b')
plt.plot(bo2.depth, dt,'r')
plt.title('de-spiked sonic')
```

# TWT-to-depth relationship

```python
dummy = np.zeros(dt.shape)
scaled_dt = 0.1524 * np.nan_to_num(dt) / 1e6
tcum = 2 * np.cumsum(scaled_dt) 
tdr = tcum + log_start_time

DT = dt **-1 *10**6 

#Redefining acoustic impedance based on the smoothed and despiked logs of vp and rho:

dt_dummy = DT[ind_] 
z = np.array(bo2.depth)
rho_dummy = rho[ind_]

ai = dt_dummy * rho_dummy
RC = (ai[1:] - ai[:-1]) / (ai[1:] + ai[:-1])

ai_fixed = dummy.copy()
ai_fixed[ind_] = ai
ai = ai_fixed


RC_fixed = dummy.copy()
RC_fixed[ind_[0][1:]] = RC
RC = RC_fixed

#plt.figure()
#plt.plot(BO2X_checkshot['MD'],BO2X_checkshot['TWT PICKED'])
#plt.title('Depth vs. TWT')
#plt.xlabel('MD [m]'), plt.ylabel('TWT [ms]')
#bo2.depth
BO2X_checkshot['TWT PICKED'].max()
```

## Converting logs to two-way-travel time

```python
# RESAMPLING FUNCTION
dt_sample = 0.004

maxt = BO2X_checkshot['TWT PICKED'].max()
#Creating TWT array:
t = np.arange(BO2X_checkshot['TWT PICKED'].loc[BO2X_checkshot.MD >= bo2.depth.min()].min()/1000., maxt/1000., dt_sample) 
# Interpolating AI from depth-log to the TWT array created above:
ai_t = np.interp(x = t, xp = tdr[ind_], fp = ai[ind_])
# Reflection coefficients in time:
RC_t = (ai_t[1:] - ai_t[:-1]) / (ai_t[1:] + ai_t[:-1])


#df.loc[(df[1].index >= 8396.83) & (df[1].index <= 8410.5)]



#BO2X_checkshot['TWT PICKED'].loc[BO2X_checkshot.MD >= bo2.depth.min()]
print(t.min(), t.max())
```

## Creating Synthetics

```python
# Define a wavelet.
def ricker(f, length, dt):
    t = np.linspace(-length / 2, (length-dt) / 2, length / dt)
    y = (1. - 2.*(np.pi**2)*(f**2)*(t**2))*np.exp(-(np.pi**2)*(f**2)*(t**2))
    return t, y


RC_t = np.nan_to_num(RC_t)
tw, w = ricker (f=25, length = 0.512, dt = 0.004)
#Convolution of Ricker wavelet with reflection coeffs in time:
synth = np.convolve(w, RC_t, mode='same')
```
